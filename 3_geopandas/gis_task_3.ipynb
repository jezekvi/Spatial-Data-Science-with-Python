```python
import geopandas as gpd
```

```python
land_cover = gpd.read_file('Land_cover.shp')
vodni_toky = gpd.read_file('Vodni_tok_HU.shp')
```

```python
print(land_cover.crs)
print(vodni_toky.crs)

land_cover = land_cover.to_crs(5514)

print(land_cover.crs)
```

```python
display(land_cover)
display(vodni_toky)
land_cover.plot()
vodni_toky.plot()
```

```python
vhodne_typy = [
  '2.3.1. Louky a pastviny',
  '2.4.2. Směsice polí, luk a trvalých plodin',
  '2.2.2. Sady, chmelnice a zahradní plantáže',
  '2.1.1. Nezavlažovaná orná půda'
]

# filtrace parcel splnujici podminky zadani
vhodny_lc = land_cover[land_cover['TRIDACORCZ'].isin(vhodne_typy)].copy()

display(vhodny_lc)
```

```python
zastavba_typy = [
  '1.1.1. Souvislá městská zástavba',
  '1.1.2. Nesouvislá městská zástavba'
]

# filtrace parcel splnujici podminky zadani
zastavba = land_cover[land_cover['TRIDACORCZ'].isin(zastavba_typy)].copy()

display(zastavba)
```

```python
# buffer

# buffer pro zastavbu
buffer_zastavba = zastavba.buffer(1000).union_all()


#buffer pro vodni toky
buffer_voda = vodni_toky.buffer(3000).union_all()
```

```python
# intersect: vodni toky a vhodna uzemi
mista_u_vodu = vhodny_lc.overlay(gpd.GeoDataFrame(geometry=[buffer_voda], crs = 5514), how = 'intersection' )

# rozdil od buffer_zastavba
finalni_vyber = mista_u_vodu.overlay(gpd.GeoDataFrame(geometry=[buffer_zastavba], crs = 5514), how = 'difference' )
```

```python
display(finalni_vyber)
```

```python
# plocha v km2

plocha_km2 = finalni_vyber.area.sum() / 1_000_000

print(f'Plocha v km2: {plocha_km2}')
```

```python
finalni_vyber.plot()
```

```python
# export

finalni_vyber.to_file("rezidence_vysledek.geojson", driver = 'GeoJSON')
```
